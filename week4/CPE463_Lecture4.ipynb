# Contrast stretching

```
# Create zeros array to store the stretched image
minmax_img = np.zeros((gray_img.shape[0],gray_img.shape[1]),dtype = 'uint8')
temp = gray_img.copy()
temp[temp > 100] = 0

# Loop over the image and apply Min-Max formulae
for i in range(temp.shape[0]):
    for j in range(temp.shape[1]):
        minmax_img[i,j] = 255*(temp[i,j]-np.min(temp))/(np.max(temp)-np.min(temp))
 
# Displat the stretched image
plt.imshow(minmax_img,cmap="gray")
```

```
plt.hist(minmax_img.ravel(),256,[0,256])
plt.title('Histogram for gray scale picture')
plt.show()
```

# Histogram Equalization

```
equ = cv2.equalizeHist(gray_img)
res = np.hstack((gray_img,equ)) #stacking images side-by-side
plt.imshow(res, cmap = "gray")
```
