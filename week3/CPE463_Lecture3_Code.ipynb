# Exercise

```
image_copy = image.copy()
```

```
image_copy[:,:,0].shape
```

```
for x_idx in range(image_copy[:,:,2].shape[0]):
  for y_idx in range(image_copy[:,:,2].shape[1]):
    if image_copy[x_idx, y_idx, 2] > 125:
      image_copy[x_idx, y_idx, 2] = 1
    else :
      image_copy[x_idx, y_idx, 2] = 0
```

```
image_copy.shape
```

```
from matplotlib.colors import hsv_to_rgb
rgb = hsv_to_rgb(image_copy)
```
